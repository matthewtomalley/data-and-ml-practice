```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from palmerpenguins import load_penguins
import scipy.stats as stats
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
data = load_penguins()
data.head()
```

```python
data.isna().any().any()
```

```python
data.info()
```

```python
data = data.dropna()
```

```python
data.describe()
```

```python
data.groupby('species').describe()
```

```python
# Body mass distrubition

adelie = data[data['species'] == 'Adelie']
chinstrap = data[data['species'] == 'Chinstrap']
gentoo = data[data['species'] == 'Gentoo']

plt.figure(figsize=(10,5))

plt.hist(adelie['body_mass_g'], alpha=0.5, color='lightgreen', label='Adelie')
plt.hist(chinstrap['body_mass_g'], alpha=0.5, color='yellow', label='Chinstrap')
plt.hist(gentoo['body_mass_g'], alpha=0.5, color='orange', label='Gentoo')

plt.xlabel('body_mass_g')

plt.legend()

plt.show()
```

```python
# Bill length distrubition

adelie = data[data['species'] == 'Adelie']
chinstrap = data[data['species'] == 'Chinstrap']
gentoo = data[data['species'] == 'Gentoo']

plt.figure(figsize=(10,5))

plt.hist(adelie['bill_length_mm'], alpha=0.5, color='lightgreen', label='Adelie')
plt.hist(chinstrap['bill_length_mm'], alpha=0.5, color='yellow', label='Chinstrap')
plt.hist(gentoo['bill_length_mm'], alpha=0.5, color='orange', label='Gentoo')

plt.xlabel('bill_length_mm')

plt.legend()

plt.show()
```

```python
# Bill depth distrubition

adelie = data[data['species'] == 'Adelie']
chinstrap = data[data['species'] == 'Chinstrap']
gentoo = data[data['species'] == 'Gentoo']

plt.figure(figsize=(10,5))

plt.hist(adelie['bill_depth_mm'], alpha=0.5, color='lightgreen', label='Adelie')
plt.hist(chinstrap['bill_depth_mm'], alpha=0.5, color='yellow', label='Chinstrap')
plt.hist(gentoo['bill_depth_mm'], alpha=0.5, color='orange', label='Gentoo')

plt.xlabel('bill_depth_mm')

plt.legend()

plt.show()
```

```python
# Flipper length distrubition

adelie = data[data['species'] == 'Adelie']
chinstrap = data[data['species'] == 'Chinstrap']
gentoo = data[data['species'] == 'Gentoo']

plt.figure(figsize=(10,5))

plt.hist(adelie['flipper_length_mm'], alpha=0.5, color='lightgreen', label='Adelie')
plt.hist(chinstrap['flipper_length_mm'], alpha=0.5, color='yellow', label='Chinstrap')
plt.hist(gentoo['flipper_length_mm'], alpha=0.5, color='orange', label='Gentoo')

plt.xlabel('flipper_length_mm')

plt.legend()

plt.show()
```
